```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

## 1.- ECONDB public API
 **Real Public Consumption** of the US Citizens

This API acquires data from the *Bureau of Economic Analysis Agency*

The url source file is 'download.bls.gov' (last updated on 2022/10/13)

```python
ECONDB = pd.read_csv('https://www.econdb.com/api/series/RPUCUS/?format=csv',                    
index_col='Date', parse_dates=['Date'])
ECONDB
```

```python
ECONDB.info()
```

```python
fig, ax = plt.subplots(figsize=(18,8))
mpl.style.use('dark_background')
ax = sns.lineplot(x="Date", y="RPUCUS", data=ECONDB )
ax.set_title('US - Real Public Consumption by Decades')
ax.set_xlabel('Year Decades')
ax.set_ylabel('RPUCUS')
plt.show()
```

## 2.- EMAIL ADRESS VERYFIER - EVA
This API validates if the given email is disposable, derivable, has a valid domain and syntax, etc.

```python
import http.client
import mimetypes
```

```python

conn = http.client.HTTPSConnection("api.eva.pingutil.com")
payload = ''
headers = {}
conn.request("GET", "/email?email=victor.torresv@my.unitec.edu.mx", payload, headers) #Verify if my school email is really a valid adress not a spam or disposable email
res = conn.getresponse()
data = res.read()
print(data.decode("utf-8"))
```
